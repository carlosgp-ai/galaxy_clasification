```
imglx1 = "2MASXJ09065870_SDSS.jpg"
imglx2 = "CGCG163-062_SDSS.jpg"
imglx3 = "CGCG242-073_SDSS.jpg"
imglx4 = "CGCG429-012_SDSS.jpg"
imglx5 = "CGCG536-030_SDSS.jpg"
imglx6 = "ESO540-G003_SDSS.jpg"
imglx7 = "IC0159_SDSS.jpg"
imglx8 = "IC0208_SDSS.jpg"
imglx9 = "IC0307_SDSS.jpg"

def imagetoarray(imglx):
  # load the image
  #image = Image.open(imglx)
  # convert image to numpy array
  galax = asarray(imglx)
  samples = expand_dims(galax, 0)
  # create image data augmentation generator
  datagen = ImageDataGenerator(zoom_range=[0.4,0.4])
  # prepare iterator
  it = datagen.flow(samples)
  # generate batch of images
  batch = it.next()
  # convert to unsigned integers for viewing
  imagereschaled = batch[0].astype('uint8')
#  print(type(galax))
  # summarize shape
 # print(galax.shape)
  
  return imagereschaled
```

```
def grayimage(img):
    imagen = Image.open(img)
    imagen = imagen.convert("RGB")
    width, height = imagen.size
    imggray = Image.new("RGB",(width,height))
    for i in range(width):
        for j in range(height):
            r,g,b = imagen.getpixel((i,j))
            average = int(floor((r+g+b)/3))                                        
            imggray.putpixel((i,j),(average,average,average))
    return imggray
```

```
imglx1 = grayimage(imglx1)
imglx2 = grayimage(imglx2)
imglx3 = grayimage(imglx3)
imglx4 = grayimage(imglx4)
imglx5 = grayimage(imglx5)
imglx6 = grayimage(imglx6)
imglx7 = grayimage(imglx7)
imglx8 = grayimage(imglx8)
imglx9 = grayimage(imglx9)
```

```
galaxia1 = imagetoarray(imglx1)
galaxia2 = imagetoarray(imglx2)
galaxia3 = imagetoarray(imglx3)
galaxia4 = imagetoarray(imglx4)
galaxia5 = imagetoarray(imglx5)
galaxia6 = imagetoarray(imglx6)
galaxia7 = imagetoarray(imglx7)
galaxia8 = imagetoarray(imglx8)
galaxia9 = imagetoarray(imglx9)

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(16,10), sharex=True, sharey=True)
ax[0,0].imshow(galaxia1)
ax[0,1].imshow(galaxia2)
ax[0,2].imshow(galaxia3)
ax[1,0].imshow(galaxia4)
ax[1,1].imshow(galaxia5)
ax[1,2].imshow(galaxia6)
ax[2,0].imshow(galaxia7)
ax[2,1].imshow(galaxia8)
ax[2,2].imshow(galaxia9)
```

```
def maxclarity(galaxia):
  max_per_row = np.max(galaxia, axis=1)
  len(max_per_row)
  # Vamos a representar ahora el m√°ximo por filas
  plt.plot(max_per_row)
  plt.show()
  return max_per_row
```

```
galaxclarity1 = maxclarity(galaxia1)
galaxclarity2 = maxclarity(galaxia2)
galaxclarity3 = maxclarity(galaxia3)
galaxclarity4 = maxclarity(galaxia4)
galaxclarity5 = maxclarity(galaxia5)
galaxclarity6 = maxclarity(galaxia6)
galaxclarity7 = maxclarity(galaxia7)
galaxclarity8 = maxclarity(galaxia8)
galaxclarity9 = maxclarity(galaxia9)

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(16,10), sharex=True, sharey=True)
ax[0,0].imshow(galaxclarity1)
ax[0,1].imshow(galaxclarity2)
ax[0,2].imshow(galaxclarity3)
ax[1,0].imshow(galaxclarity4)
ax[1,1].imshow(galaxclarity5)
ax[1,2].imshow(galaxclarity6)
ax[2,0].imshow(galaxclarity7)
ax[2,1].imshow(galaxclarity8)
ax[2,2].imshow(galaxclarity9)
```
